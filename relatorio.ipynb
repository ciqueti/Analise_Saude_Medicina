```python
# Selecionado menos labels para que nosso xticks não fique polúido visualmente
# Função recebe ticks em datas e espaçamento, e retorna ticks e labels para a função xticks
def x_indices(indices, espacamento):
    ticks = []
    label = []
    cont = 0
    for i in indices:
        if (cont%espacamento == 0):
            ticks.append(cont)
            label.append(i)
        cont += 1
    return(ticks,label)
```

```python
fig, ax = plt.subplots(figsize=(12,5))
plt.xticks(x_indices(confirmed.loc['China'][2:].index, 4)[0], x_indices(confirmed.loc['China'][2:].index, 4)[1])
ax.plot(confirmed.loc['China'][2:])
ax.set_title('Casos Confirmados na China', fontsize=15)
ax.set_xlabel('Data', fontsize=12)
ax.set_ylabel('Quantidade', fontsize=12)
```

## 3 Fonte de dados e letalidade
### 3.1 Carregando deaths, recovered

```python
deaths = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
recovered = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
confirmed = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
```

```python
for i in recovered.index:
    for k in recovered.iloc[:,4:].columns:
        if recovered.loc[i,k] != 0:
            recovered.loc[i,'Total Recovered'] = recovered.loc[i,k]

#Por algum motivo os recovered passam a ser 0 a partir de alguma data, sendo assim, usamos o último resultado disponível
```

```python
recovered
```

```python
recovered[['Country/Region', 'Total Recovered']].groupby(by=['Country/Region']).sum()
```

### 3.2 Criando Dataframe com as 3 taxas por país do último dia disponível

```python
def latest_by_country(data):
    latest = data.groupby(by=['Country/Region']).sum().iloc[:,-1]
    return(latest)
```

```python
latest_cases = [latest_by_country(confirmed), latest_by_country(deaths), recovered[['Country/Region', 'Total Recovered']].groupby(by=['Country/Region']).sum()]
latest_cases
```

```python
combined = pd.concat(latest_cases, axis='columns')
combined.columns = ['Confirmed', 'Deaths', 'Recovered']
combined
```

```python
letalidade = combined.sum()['Deaths'] / combined.sum()['Confirmed']
print(f'Taxa de Letalidade = {letalidade*100} %')

#Esse número representa uma estimativa calculada a partir dos dados, porém temos vários outros elementos como fonte dos dados, formalização dos testes de cada país, dentre outos para obtermos o valor preciso.
```

```python
letalidade2 = combined['Deaths'].sum() / (combined['Deaths'].sum() + combined['Recovered'].sum())
print(f'Taxa de Letalidade 2 = {letalidade2*100} %')
```

```python
un_population_total = tota_population.query('Year == 2018')[['Country Name', 'Value']].set_index('Country Name')
un_population_total['Country_Name'] = un_population_total.index
un_population_total.head()
```

US ; Korea, South  ; Russia	 - Possuem nomes distintos em cada dataset. Como podemos unir essas informações?

```python
un_population_total.query('Country_Name.str.contains("Korea")')
```

```python
def rename_location(location):
    if location =='Russian Federation':
        return('Russia')
    elif location == 'United States':
        return('US')
    elif location == 'Korea, Rep.':
        return('Korea, South')
    else:
        return(location)
    
```

```python
new_index = un_population_total.index.map(rename_location)
```

```python
un_population_total.set_index(new_index, inplace=True)
un_population_total
```

```python
combined = combined.join(un_population_total['Value']).sort_values(by=['Confirmed'], ascending=False).head(10)
```

```python
combined['PopTotal'] = combined['Value']
combined.drop('Value', axis='columns', inplace=True)
combined.head(10)
```

### 5.3 Taxa de Incidência e Mortalidade por 100k de cada país

```python
combined['incidence_ratio'] = (combined['Confirmed'] / combined['PopTotal']) * 100000
combined['mortality_rate'] = (combined['Deaths'] / combined['PopTotal']) * 100000
```
